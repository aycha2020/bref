# Importation et exploration

```python
import pandas as pd
```

```python
df = pd.read_csv("store1.csv")

df.info()
```

# Nettoyage

```python
df.duplicated()

# Corriger les types de colonnes (dates, numériques, catégories)

df["Order Date"] =  pd.to_datetime(df["Order Date"],format="%d/%m/%Y")
df["Ship Date"] = pd.to_datetime(df["Ship Date"],format="%d/%m/%Y")
df["Postal Code"] = df["Postal Code"].astype("Int64")

df = df.dropna()

df.info()

```

```python
df.isnull().sum()
```

```python
coln_categorie = [
"Ship Mode","Segment","Country",
"City","State","Region","Category","Sub-Category"
]
for col in coln_categorie:
    df[col]=df[col].astype('category')
print(df.dtypes)
```

# Transformation et feature engineering

```python
# Extraire des informations temporelles : année, mois, trimestre

df["Annee"]= df["Order Date"].dt.year
df["Mois"]= df["Order Date"].dt.month
df["Trimestre"]= df["Order Date"].dt.quarter
df
```

```python
# les métriques:

df["cout"]= df["Sales"]*0.8
df["profit"]= df["Sales"]-df["cout"]
df["taux_profit"]= df["profit"]/df["Sales"]*100
df["ratio"]= df["profit"]/df["Sales"]
df
```

```python
from utils import segmentation


df["segmentation"] = df["Sales"].map(segmentation)
df
```

```python
# Délais de livraison:

df["dliv"]= (df["Ship Date"]-df["Order Date"]).dt.days
df
```

```python
# Ventes par catégorie/ région:

vent_par_catg= df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
vent_par_rég= df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
print(vent_par_catg)
print(vent_par_rég)
```
